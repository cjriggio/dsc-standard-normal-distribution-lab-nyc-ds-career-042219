## Define the interval bounds where the contains 99% of the observations (hint: $3\sigma$)

```python
# Perform any calculations necessary here
mean = df.mean()
std = df.std()
```

```python
mean
```

```python
std
```

```python
# Write your answer here 
print(mean-std)
print(mean+std)
#68% of yields are between 36. 4 and 48.41
```

```python
print(mean-(std*2))
print(mean+(std*2))
#95% of yields are between 30.4 and 54.4
```

```python
print(mean-(std*3))
print(mean+(std*3))
#99% of yields are between 24.4 and 60.4
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
z = (35-mean)/std
z
```

```python
# Interpret the result
#1.23 standard deviations below the mean 
```

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

```python
# Interpret the z score
#the tree has a yield that is 1.85 stdev's higher than the mean 
```

```python
# Calculate yield
X = (1.85*std) + mean
```

```python
# What is the yield ?
X #53.5 pounds
```
